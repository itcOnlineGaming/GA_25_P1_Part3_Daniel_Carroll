```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data/player_logged_in.csv')

#Converts the time column to datetime
data['Time'] = pd.to_datetime(data['Time'])

#Extract just the date part
data['Date'] = data['Time'].dt.date


#Groups by date and counts unique player IDs
daily_active_players = data.groupby('Date')['pid'].nunique().reset_index()
daily_active_players.columns = ['Date', 'UniqueActivePlayers']

#Sort by date for plotting
daily_active_players.sort_values('Date', inplace=True)



#Line Graph
plt.figure(figsize=(14, 6))
plt.plot(daily_active_players['Date'], daily_active_players['UniqueActivePlayers'], marker='o')
plt.title('Daily Unique Active Players (Line Graph)')
plt.xlabel('Date')
plt.ylabel('Number of Unique Players')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()


#Bar Graph
plt.figure(figsize=(14, 6))
plt.bar(daily_active_players['Date'], daily_active_players['UniqueActivePlayers'])
plt.title('Daily Unique Active Players (Bar Chart)')
plt.xlabel('Date')
plt.ylabel('Number of Unique Players')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
